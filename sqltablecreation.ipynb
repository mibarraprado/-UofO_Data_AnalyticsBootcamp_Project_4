```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#importing CSV files
test_df = pd.read_csv('resources/test.csv')
train_df = pd.read_csv('resources/train.csv')
```

```python
#checking the data frame
test_df.head()
```

```python
test_df.head()
```

```python
#checking the data frame
train_df.head()
```

```python
#combining the two datasets
combined_df = pd.concat([train_df, test_df], axis=0, sort=False)
combined_df.head()

```

```python
# counting the rows in each dataframe to make sure the data is correct
train_count = len(train_df)
test_count =  len(test_df)
combined_count = len(combined_df)

print(f'Train count: {train_count}, Test count: {test_count}, Combined count:{train_count+test_count} Combined_df count: {combined_count}')
```

```python
# checking the data types of the columns
combined_df.value_counts('Transported')
```

```python
#count the nan values in the Transported column
combined_df['Transported'].isna().sum()
```

```python
#checking the data types
combined_df.dtypes
```

```python
#converting cryosleep, VIP and Transported into boolean values
combined_df['CryoSleep'] = combined_df['CryoSleep'].astype(bool)
combined_df['VIP'] = combined_df['VIP'].astype(bool)


```

```python
combined_df.dtypes
```

```python
#creating passenger table
passenger_df = combined_df[['PassengerId', 'Name', 'HomePlanet', 'Cabin', 'Age', 'Transported']]
passenger_df.head()
```

```python
#creating a planet table
planet_df = combined_df[['PassengerId','HomePlanet','Destination', 'CryoSleep']]
planet_df.head()
```

```python
#creating an amenities table
amenities_df = combined_df[['PassengerId','Cabin','VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']]
amenities_df.head()
```

```python
#exporting the dataframes to csv files with a header
passenger_df.to_csv('resources/passenger.csv', index=False, header=True)
planet_df.to_csv('resources/planet.csv', index=False, header=True)
amenities_df.to_csv('resources/amenities.csv', index=False, header=True)
```
